```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

recent_grads = pd.read_csv('recent-grads.csv')
recent_grads.iloc[0]
```

```python
recent_grads.head(5)
```

```python
recent_grads.tail(5)
```

```python
recent_grads.describe()
```

```python
raw_data_count = len(recent_grads)
recent_grads = recent_grads.dropna()
cleaned_data_count = len(recent_grads)
print("Raw count: {}\nCleaned count: {}".format(raw_data_count, cleaned_data_count))
```

```python
ax = recent_grads.plot(x='Sample_size', y='Median', kind='scatter')
ax.set_title('Sample Size vs Median Income')
```

```python
ax = recent_grads.plot(x='Sample_size', y='Unemployment_rate', kind='scatter')
ax.set_title('Sample Size vs Unemployment Rate')
```

```python
ax = recent_grads.plot(x='Full_time', y='Median', kind='scatter')
ax.set_title('Number of Full-Time Employees vs Median Income')
```

```python
ax = recent_grads.plot(x='ShareWomen', y='Unemployment_rate', kind='scatter')
ax.set_title('% Share of Women vs Unemployment Rate')
```

```python
ax = recent_grads.plot(x='Men', y='Median', kind='scatter')
ax.set_title('Number of Men vs Median Income')
```

```python
ax = recent_grads.plot(x='Women', y='Median', kind='scatter')
ax.set_title('Number of Women vs Median Income')
```

```python
cols = ['Sample_size', 'Median', 'Employed', 'Full_time', 'ShareWomen', 'Unemployment_rate', 'Men', 'Women']
fig = plt.figure(figsize=(6,24))
for i in range(8):
    ax = fig.add_subplot(8,1,i+1)
    ax = recent_grads[cols[i]].plot(kind='hist',rot=40)
    ax.set_title(cols[i])
```

```python
from pandas.tools.plotting import scatter_matrix
scatter_matrix(recent_grads[['Sample_size','Median']], figsize=(10,10))
```

```python
scatter_matrix(recent_grads[['Sample_size','Median','Unemployment_rate']],figsize=(10,10))
```

```python
ax = recent_grads.head(10).plot.bar(x='Major', y='ShareWomen', legend=False)
ax.set_title('% Share of Women per 10 Highest Paying Majors')
```

```python
ax = recent_grads.tail(10).plot.bar(x='Major', y='ShareWomen', legend=False)
ax.set_title('% Share of Woman per 10 Lowest Paying Majors')
```

```python
ax = recent_grads.head(10).plot.bar(x='Major',y='Unemployment_rate',legend=False)
ax.set_title('Unemployment Rate for 10 Highest Paying')
```
